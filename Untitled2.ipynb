## **ÁLGEBRA Y COMPOSICIÓN DE FUNCIONES**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las funciones
def f(x):
    return 2*x

def g(x):
    return x**2

# Definir el rango de valores de x
x = np.linspace(-3, 3, 400)

# Calcular los valores de y para cada función
y_f = f(x)
y_g = g(x)
y_h = y_f + y_g

# Graficar las funciones
plt.figure(figsize=(8, 6))
plt.plot(x, y_f, label='f(x) = 2x ', color='black')
plt.plot(x, y_g, label='g(x) = x^2', color='blue')
plt.plot(x, y_h, label='h(x) = f(x) + g(x)', color='red')

# Etiquetas y leyenda
plt.xlabel('x')
plt.ylabel('y')
plt.title('Suma de funciones')
plt.legend()

# Mostrar la gráfica
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las funciones
def f(x):
    return 2*x

def g(x):
    return x**2 + 1 # Aseguramos que g(x) nunca sea igual a cero

# Definir el rango de valores de x
x = np.linspace(-3, 3, 400)

# Calcular los valores de y para cada función
y_f = f(x)
y_g = g(x)

# Calcular el cociente de las funciones
y_quotient = np.divide(y_f, y_g, out=np.zeros_like(y_f), where=y_g!=0)  # Evitamos la división por cero

# Graficar las funciones
plt.figure(figsize=(8, 6))
plt.plot(x, y_f, label='f(x) = 2x', color='black')
plt.plot(x, y_g, label='g(x) = x^2 + 1', color='blue')
plt.plot(x, y_quotient, label='f(x) / g(x)', color='red')

# Etiquetas y leyenda
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cociente de funciones')
plt.legend()

# Mostrar la gráfica
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las funciones
def f(x):
    return x**2

def g(x):
    return 2*x

# Definir el rango de valores de x
x = np.linspace(-3, 3, 400)

# Calcular la composición de las funciones
y_composition = f(g(x))

# Graficar la composición de funciones
plt.figure(figsize=(8, 6))
plt.plot(x, y_composition, label='f(g(x)) = (2x)^2 = 4x^2', color='red')

# Etiquetas y leyenda
plt.xlabel('x')
plt.ylabel('y')
plt.title('Composición de funciones ')
plt.legend()

# Mostrar la gráfica
plt.grid(True)
plt.show()
```

## **FUNCIÓN EXPONENCIAL**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función exponencial con base mayor a 1
def f(x, base):
    return base ** x

# Definir el rango de valores de x
x = np.linspace(-3, 3, 400)

# Definir las bases de las funciones exponenciales (mayor a 1)
bases = [2]

# Graficar las funciones exponenciales
plt.figure(figsize=(8, 6))
for base in bases:
    y = f(x, base)
    plt.plot(x, y, label=f'y = {base}^x')

# Etiquetas y leyenda
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Función exponencial con base {base}')
plt.legend()

# Mostrar la gráfica
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función exponencial con base entre 0 y 1
def f(x, base):
    return base ** x

# Definir el rango de valores de x
x = np.linspace(-4, 4, 400)

# Definir las bases de las funciones exponenciales (entre 0 y 1)
bases = [0.5]

# Graficar las funciones exponenciales
plt.figure(figsize=(8, 6))
for base in bases:
    y = f(x, base)
    plt.plot(x, y, label=f'y = {base}^x')

# Etiquetas y leyenda
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Función exponencial con base {base}')
plt.legend()

# Mostrar la gráfica
plt.grid(True)
plt.show()

```
